```python
WEBSITES = {
    "alcamp": "./data/maps/alcamp.json"
}
```

```python
class Engine:
    def __init__(self, website):
        with open(WEBSITES[website], "r") as f:
            data = f.read()
        self.website_map = Website.model_validate_json(data)

    def welcome_message(self):
        return f"""
Greet the user and offer your assistance. You should start by speaking in Arabic.
and offer the user available options ( flows ) to start/select
here is the flows desciption so user can choose what he wants to do :

{self.website_map.flows_desciption}
"""
```

```python
e = Engine("alcamp")
```
